```python
!pip install tiktoken
import tiktoken
enc = tiktoken.get_encoding("cl100k_base")
```

```python
text1 = "hello"
text2 = "Hello"
text3 = "नमस्ते"
text4 = "1234567890"
text5 = "strawberry"

for text in [text1, text2, text3, text4, text5]:
    tokens = enc.encode(text)
    print(f"{text:15} → {tokens}")
```

```python
def show_tokens(text):
    tokens = enc.encode(text)
    decoded = [enc.decode([t]) for t in tokens]
    print(f"Text: {text}")
    print(f"Tokens: {tokens}")
    print(f"Decoded chunks: {decoded}")
    print(f"Token count: {len(tokens)}")
    print()

show_tokens("hello world")
show_tokens("नमस्ते दुनिया")
show_tokens("98 + 57 = ?")
show_tokens("strawberry")
```

```python
show_tokens("1234 + 5678 = ?")
show_tokens("99999 + 11111 = ?")
```

```python
texts = {
    "English": "I love machine learning",
    "Hindi": "मुझे मशीन लर्निंग पसंद है",
    "Numbers": "1234 + 5678 = 6912",
}

for lang, text in texts.items():
    tokens = enc.encode(text)
    print(f"{lang:10} → {len(tokens)} tokens for {len(text)} characters")
```
